```python
df = pd.read_excel(excel_file, Sheet_name='Sheet1')
print('조회 건수 : %d건' % df.shape[0])
print(df['상품명'].unique())
df.head(3)
```

```python
read_columns = ['주담당채널', '영업채널명', '영업담당자명', '고객명', '사업자/주민', '주문일자', '주문상태', '주문현황', '납입구분', '상품명', '단위상품명', '공급가', '금액(VAT별도)', '판매유형', '기회시작일' ]
df2 = df[read_columns]
print('조회 건수 : %d건' % df.shape[0])
df2.head(3)
```

```python
# 상품명
product_category = {'ICUBE 클라우드서버', 'ICUBE/IU 클라우드서버', 'ICUBE/GW 클라우드서버', 'IU 클라우드서버'
                    , 'IU/GW 클라우드서버', 'GW 클라우드서버', 'Private Cloud', 'iU Private', 'iCUBE Private'
                    , '클라우드서버', '클라우드백업서버', 'Bizbox Alpha Cloud','(신)iCUBE Cloud Edition'}
#product_category = {'GW 클라우드서버', 'ICUBE/GW 클라우드서버', 'IU/GW 클라우드서버'}  # 집합(중복값 불허)
df3 = df2[df2['상품명'].isin(product_category)]
print('조회 건수 : %d건' % df3.shape[0])
print(df3['상품명'].value_counts())
df3.head(3)
```

```python
df3[df3['상품명']=='ICUBE/IU 클라우드서버']
```

```python
# group by
groupby_columns_list = ['영업채널명', '영업담당자명', '고객명', '사업자/주민', '주문일자', '상품명', '판매유형']
grouped = df3.groupby(groupby_columns_list)

df4 = grouped.sum().reset_index()
print(df4.columns)
print('조회 건수 : %d건' % df4.shape[0])
print(df4['상품명'].value_counts())
"""
# column rename
#df6.rename(columns={'NEW_NAME':'OLD_NAME'}, inplace=True)   # 속도 느림
df6.columns = ['CNT' if x == 0 else x for x in df6.columns]    # 속도 빠름
print(df6.columns)

# sort
df6.sort_values(by=['주문일자', 'CNT'], ascending=[False, True], inplace=True)
print('조회 건수 : %d건' % df6.shape[0])
df6.head(3)
"""
df4
```

```python
# IT코디센터는 C로 대리점은 B로 새로운칼럼에 추가한다
channel_series = df4.apply(lambda v: 'C' if ((v[0].endswith('IT코디센터'))|(v[0]=='강원지점')|(v[0]=='제주지점')) else 'B', axis=1)   #
#df4['CB'] = channel_series
df4.loc[:, 'C_B'] = channel_series
print('조회 건수 : %d건' % df4.shape[0])
print(df4['C_B'].value_counts())
df4
```

```python
# sort
df5 = df4.sort_values(by=['C_B', '영업채널명', '영업담당자명'], ascending=[False, True, False])
print('조회 건수 : %d건' % df5.shape[0])
df5.head(3)
```

```python
# filter
filter_product = {'ICUBE 클라우드서버', 'ICUBE/IU 클라우드서버', 'ICUBE/GW 클라우드서버', 'IU 클라우드서버'
                    , 'IU/GW 클라우드서버', 'GW 클라우드서버'}

df6 = df5[df5['상품명'].apply(lambda x : x in filter_product)]
print('조회 건수 : %d건' % df6.shape[0])
df6
```

```python
# 업체명/솔루션/년금액/센터/담당자/판매유형
for row in df6.values:
    product_price = "{:,}원".format(int(row[8]))
    order_type = '재계약' if row[6] == '기타' else row[6]
    print("- {0}/{1}/년{2}원/{3}/{4}/{5}".format(row[2], row[5].split()[0], product_price, row[0].split()[0], row[1], order_type))
    
```

```python
import re
def commaParse(num):
    return re.sub('(?<=\d)(?=(\d{3})+(?!\d))',',',str(num))
```
