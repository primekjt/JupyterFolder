```python
import openpyxl
#import locale
import re #regex
from datetime import datetime

def commaParse(num):
    return re.sub('(?<=\d)(?=(\d{3})+(?!\d))',',',str(num))

DATA_PATH = "C:\\DEV\\DATA\\"
```

```python
# 엑셀파일 열기
#excel_file = "C:\\DEV\\DATA\\클라우드사업부 미수리스트_2019.xlsx"
excel_file = "C:\\DEV\\DATA\\Sheet1.xlsx"

wb = openpyxl.load_workbook(excel_file)
 
```

```python
# 현재 Active Sheet 얻기
#ws = wb.active
# 선택 Sheet 얻기
#ws = wb.get_sheet_by_name("Sheet1")
ws = wb["Sheet1"]
```

```python
# 국영수 점수를 읽기
for r in ws.rows:
    #row_index = r[0].row   # 행 인덱스
    company = r[3].value
    money = commaParse(r[9].value)
    itcoodi = r[0].value
    name = r[1].value
    
    #locale.setlocale(locale.LC_ALL, '')  # 시스템 기본 로케일 사용
    #money = local.format('%f', float(r[11].value), 1)

    
    out_text = """
[장기 미수 안내 및 입금일자 확인 요청 ] {0} : 총미수금 {1}원(VAT포함가) 
수신 : {2} {3}
안녕하십니까? 클라우드사업부 김진태 부장입니다. 

장기 미수금 발생 건으로 해당 고객사에 수금일자/진행사항 확인 요청드리며,

가능한 빠른 입금 일정 확인 후 전체 회신 부탁드립니다. (전체 회신 요망)

▶ 2019년 04월 18~19일 고객사 안내 메일 발송 완료
▶ {0} : 총미수금 {1}원(VAT포함가)

※ 추후 입금일자 미회신 또는 대금결재가 이루어지지 않을 경우
    서버 종료후 채권추심 진행예정이오니 참고 부탁드립니다. 
※ 장기 미수 고객사는 서비스 중지 및 채권 추심 진행할 수 있사오니, 참조 바랍니다.

감사합니다.
"""
    # 합계 쓰기
    #ws.cell(row=row_index, column=5).value = sum
 
    print(out_text.format(company, money, itcoodi, name))
    print('-'*50)
```
